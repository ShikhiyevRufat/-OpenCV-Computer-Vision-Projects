```python
new_image = cv2.imread('images/lake.jpg', cv2.IMREAD_COLOR)
b, g, r = cv2.split(new_image)

plt.figure(figsize=[20,5])
plt.subplot(1,4,1);plt.imshow(b, cmap="gray");plt.title("blue")
plt.subplot(1,4,2);plt.imshow(g, cmap="gray");plt.title("green")
plt.subplot(1,4,3);plt.imshow(r, cmap="gray");plt.title("red")

img_merge = cv2.merge([b,g,r])
plt.subplot(1,4,4);plt.imshow(img_merge[:, :, ::-1]);plt.title("orginal")
```

```python
new_imagehsv = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)
new_imagehsv = cv2.cvtColor(new_imagehsv, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(new_imagehsv)

plt.imshow(new_imagehsv)

plt.figure(figsize=[20,5])
plt.subplot(1,4,1);plt.imshow(h, cmap="gray");plt.title("blue")
plt.subplot(1,4,2);plt.imshow(s, cmap="gray");plt.title("green")
plt.subplot(1,4,3);plt.imshow(v, cmap="gray");plt.title("red")

img_merge = cv2.merge([h,s,v])
plt.subplot(1,4,4);plt.imshow(img_merge[:, :, ::-1]);plt.title("orginal")
```

```python
cv2.imwrite("images/My_new_writeen_image.png", img_merge )
```
