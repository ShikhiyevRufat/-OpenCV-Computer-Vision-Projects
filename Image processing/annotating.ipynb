```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
Image('images/apolla.jpg')
```

```python
img = cv2.imread('images/apolla.jpg')
img
```

```python
plt.imshow(img)
```

```python
cv2.line(img, (200,100), (400,100), (0,0,255), 5)
plt.imshow(img)
```

```python
cv2.circle(img, (800,500), 100, (200,0,0), 5)
plt.imshow(img)
```

```python
cv2.rectangle(img, (420,150), (600, 570), (0,200,0), 5)
plt.imshow(img)
```

```python
cv2.putText(img, "Welcome to Apolla 11", (200,800), cv2.FONT_HERSHEY_SIMPLEX, 2.3, (0,255,0), 5)
plt.imshow(img)
```
