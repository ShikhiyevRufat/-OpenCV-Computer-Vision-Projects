```python
import cv2
import math
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imgs = sorted(glob.glob('panaroma_img/*.jpg'))
imgs = [cv2.cvtColor(cv2.imread(f), cv2.COLOR_BGR2RGB) for f in imgs]

column = 3
rows = math.ceil(len(imgs) / column)

plt.figure(figsize=[30,10])

for i, imgs_file in enumerate(imgs):
    plt.subplot(rows, column, i+1)
    plt.axis("off")
    plt.imshow(imgs_file)

plt.show()    
```

```python
stitcher = cv2.Stitcher_create()
status, result = stitcher.stitch(imgs)

if status == 0:
    plt.figure(figsize=[30, 10])
    plt.imshow(result)
```
