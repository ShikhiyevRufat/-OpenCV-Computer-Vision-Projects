```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
import cv2
```

```python
img = cv2.imread("images/checkerboard_84x84.png", 0)
print("Image shape:", img.shape)
print("Image dtype:", img.dtype)

plt.imshow(img)
```

```python
img_width = 10
img_height = 10
dsize = (img_width, img_height)
resize_img = cv2.resize(img, dsize)
plt.imshow(resize_img)
```

```python
resize_img
```

```python
resize_img[0, 1] = 200
resize_img[0, 2] = 200
resize_img[1, 1] = 200
resize_img[1, 2] = 200

resize_img
plt.imshow(resize_img, cmap="gray")
```

```python
img2 = cv2.imread("images/coca-cola-logo.png", cv2.IMREAD_COLOR)
plt.imshow(img2)
```

```python
img_flip1 = cv2.flip(img2, 0)
img_flip2 = cv2.flip(img2, 1)
img_flip3 = cv2.flip(img2, -1)
img_flip4 = img2

plt.figure(figsize=[25,5])
plt.subplot(1,4,1);plt.imshow(img_flip1);plt.title("Horizontal")
plt.subplot(1,4,2);plt.imshow(img_flip2);plt.title("Vertical")
plt.subplot(1,4,3);plt.imshow(img_flip3);plt.title("Horizontal and Vertical")
plt.subplot(1,4,4);plt.imshow(img_flip4);plt.title("Orginal")
```
