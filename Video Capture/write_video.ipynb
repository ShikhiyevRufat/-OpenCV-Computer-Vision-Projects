```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
source = 'video/test1.mp4'
cap = cv2.VideoCapture(source)
```

```python
if not cap.isOpened():
    print("Error opening file or your source is false")
```

```python
ret, frame = cap.read()

plt.imshow(frame[:, :, ::-1])
```

```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

out_avi = cv2.VideoWriter("testing1.avi", cv2.VideoWriter_fourcc("M", "J", "P", "G"), 10, (frame_width, frame_height))
out_mp4 = cv2.VideoWriter("testing1.mp4", cv2.VideoWriter_fourcc(*"XVID"), 10, (frame_width, frame_height))
```

```python
while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        out_avi.write(frame)
        out_mp4.write(frame)
    else:
        break
```

```python
cap.release()
out_avi.release()
out_mp4.release()
```
